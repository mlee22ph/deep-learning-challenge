## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(["EIN", "NAME"], axis=1)
application_df.head()
```

```python
# Determine the number of unique values in each column.
column_list = application_df.dtypes.index.tolist()
application_df[column_list].nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = application_df["APPLICATION_TYPE"].value_counts()
application_type_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = []
for indx, values in application_type_counts.items():
  if (values < 500):
    application_types_to_replace.append(indx)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df["CLASSIFICATION"].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts.loc[lambda x: x>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = []
for indx, values in classification_counts.items():
  if (values < 1000):
    classifications_to_replace.append(indx)


# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Just checking that the new classification 'Other' is present
application_df.head(20)
```

```python
# Check data types
# Categorical data, which are the object types, needs some more preprocessing
application_df.dtypes
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummy_result = pd.get_dummies(application_df.drop(columns=['STATUS', 'ASK_AMT', 'IS_SUCCESSFUL']))
dummy_result.head()
```

```python
# Split our preprocessed data into our features and target arrays
# Save target info
target_df = application_df['IS_SUCCESSFUL']
target_df.head()
```

```python
# Save features info
features_df = dummy_result
features_df[['STATUS', 'ASK_AMT']] = application_df[['STATUS', 'ASK_AMT']]

features_df.head()
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(features_df, target_df, random_state=42, stratify=target_df)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Increased number of nuerons from 50 to 100
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input_dimentions = 43
number_of_units = 100

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=number_of_units, activation="relu", input_dim=input_dimentions))

# Second hidden layer
# Increaded number of nuerons from 30 to 50
nn.add(tf.keras.layers.Dense(units=number_of_units, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
